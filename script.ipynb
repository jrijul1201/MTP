```python
import os
import pandas as pd
folder_name = input("Enter the name of the folder: ")
```

```python
import matplotlib.pyplot as plt
def readDatFileAndConvertToDataFrame(path):
    baseName = os.path.basename(path).split('.')
    if(baseName[1] != "dat"):
        return
    df = pd.read_csv(path, delimiter=' ', header=None)
    plt.plot(df[0], df[1], linestyle='-')
    plt.xlabel('time for last 25 sec')
    plt.ylabel('cwnd size')
    plt.title(baseName[0])
    plt.savefig(os.path.dirname(path)+'/'+baseName[0]+'.png')
    plt.show()

def readDatFileAndConvertToDataFrameQueueSize(path):
    baseName = os.path.basename(path).split('.')
    if(baseName[1] != "dat"):
        return
    df = pd.read_csv(path, delimiter=' ', header=None)
    plt.plot(df[0], df[1], linestyle='-')
    plt.xlabel('Queue size')
    plt.ylabel('Time')
    plt.title(baseName[0])
    plt.savefig(os.path.dirname(path)+'/'+baseName[0]+'.png')
    plt.show()
```

```python
absolute_path = "/home/dipesh/ns-allinone-3.36.1/ns-3.36.1/results/" + folder_name

cwnd_folder = absolute_path + "/cwndTraces"

for filename in os.listdir(cwnd_folder):
    file_path = os.path.join(cwnd_folder, filename)
    readDatFileAndConvertToDataFrame(file_path)

readDatFileAndConvertToDataFrameQueueSize(absolute_path+"/queue-size.dat")
```
